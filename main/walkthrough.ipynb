# 1. Primeros pasos

El objetivo de este notebook es servir como guía de la implementación del Algoritmo Genético (AG) para el proyecto de la asignatura, así como la revisión de resultados obtenidos y una serie de evaluaciones sobre los distintos hiperparámetros que contiene.

## Inicialización del Algoritmo Genético

El primer paso para trabajar con el AG será la importación de la clase que lo define. Posteriormente, lo inicializamos; para ello requerimos de una serie de atributos, entre ellos una semilla para poder reproducir los resultados aquí obtenidos.

## Trabajo con Datasets

Para la ilustración del algoritmo, trabajaremos con dos datasets, 'toy' y 'housing'.


```python
from AG_class import AG
import os

nombre_dataset = 'toy1'

data_folder = "./data/"

nombre_dataset_train = os.path.join(data_folder, nombre_dataset + "_train.csv")
nombre_dataset_val = os.path.join(data_folder, nombre_dataset + "_val.csv")

ag = AG(
    datos_train=nombre_dataset_train,
    datos_test=nombre_dataset_val,
    seed=123,
    nInd=500,
    maxIter=120
)
```

nInd representará el tamaño de la población con la que trabajaremos y maxIter el número de generaciones que correrá el algoritmo. Una vez definida la clase, simplemente tenemos que ejecutar la función run() del AG. Haremos uso de el módulo time para poder medir cómo de rápida es la ejecución del AG para el problema dado.

```python
import time

inicio = time.time()
ind, y_pred = ag.run()
fin = time.time()
print(f'Tiempo ejecucion: {(fin-inicio):.2f} segundos')
```

Una vez hemos finalizado, tendremos un mejor individuo obtenido a través de las iteraciones del AG. En nuestro caso, el individuo representa los coeficientes, exponentes y la constante C que han obtenido un mejor fitness para el dataset concreto. Evaluamos ahora su capacidad predictiva con las métricas propuestas

```python
from sklearn.metrics import r2_score,root_mean_squared_error
import pandas as pd

# Imprimir mejor solución encontrada
print(f'Mejor individuo: {ind}')

# Imprimir predicciones sobre el conjunto de test
print(f'Predicciones: {y_pred}')

# Cargar valores reales de 'y' en el conjunto de validacion/test 
# y calcular RMSE y R2 con las predicciones del AG
y_true = pd.read_csv(nombre_dataset_val)['y']  # Cambiar 'target' por 'y'
rmse = root_mean_squared_error(y_true, y_pred)
print(f'RMSE: {rmse:.4f}')

r2 = r2_score(y_true, y_pred)
print(f'R2: {r2:.4f}')
```

Los valores de RMSE (Root Mean Squared Error) de 0.0664 y un coeficiente R2 de 0.9979 indican un desempeño muy alto del modelo de regresión. El RMSE bajo sugiere que las predicciones del modelo están, en promedio, muy cerca de los valores reales, implicando pocos errores de predicción. Por otro lado, un R2 de 0.9979 muestra que casi toda la variabilidad de la variable dependiente ha sido capturada por el modelo.
